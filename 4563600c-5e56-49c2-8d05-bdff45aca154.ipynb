```
# Additional analysis: Statistical test
import scipy.stats as stats

# Prepare data for statistical test
faint = df[df['Fluorescence'] == 'Faint']['Price']
medium = df[df['Fluorescence'] == 'Medium']['Price']
strong = df[df['Fluorescence'] == 'Strong']['Price']

# Perform one-way ANOVA test
f_statistic, p_value = stats.f_oneway(faint, medium, strong)

print("\nOne-way ANOVA Test Results:")
print(f"F-statistic: {f_statistic:.4f}")
print(f"p-value: {p_value:.4f}")

# Calculate percentage price difference
base_price = df[df['Fluorescence'] == 'Faint']['Price'].mean()
medium_diff = ((df[df['Fluorescence'] == 'Medium']['Price'].mean() - base_price) / base_price) * 100
strong_diff = ((df[df['Fluorescence'] == 'Strong']['Price'].mean() - base_price) / base_price) * 100

print("\nPercentage Price Difference (compared to Faint fluorescence):")
print(f"Medium Fluorescence: {medium_diff:.2f}%")
print(f"Strong Fluorescence: {strong_diff:.2f}%")

# Distribution of fluorescence by type
fluorescence_dist = pd.crosstab(df['Fluorescence'], df['Type'], normalize='columns') * 100
print("\nDistribution of Fluorescence by Type (%):")
print(fluorescence_dist)
```

5. Cut Quality and Proportions
Explore how cut quality (e.g., Ideal, Excellent) interacts with depth %, table %, and symmetry to influence price. Certain combinations of these proportions may yield higher prices.

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create visualization for price distribution by cut
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Cut', y='Price')
plt.title('Price Distribution by Cut Grade')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Create scatter plots for depth and table percentages
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

sns.scatterplot(data=df, x='Depth %', y='Price', hue='Cut', alpha=0.5, ax=ax1)
ax1.set_title('Price vs Depth % by Cut')

sns.scatterplot(data=df, x='Table %', y='Price', hue='Cut', alpha=0.5, ax=ax2)
ax2.set_title('Price vs Table % by Cut')

plt.tight_layout()
plt.show()

# Calculate average prices and counts by cut
cut_stats = df.groupby('Cut').agg({
    'Price': ['mean', 'count'],
    'Depth %': 'mean',
    'Table %': 'mean'
}).round(2)

print("\nCut Grade Statistics:")
print(cut_stats)

# Create optimal ranges analysis
def get_range_category(value, ranges):
    for range_name, (lower, upper) in ranges.items():
        if lower <= value <= upper:
            return range_name
    return 'Outside Range'

# Define ideal ranges based on GIA standards
depth_ranges = {
    'Ideal': (61.0, 62.5),
    'Excellent': (59.0, 62.9),
    'Very Good': (58.0, 64.0)
}

table_ranges = {
    'Ideal': (53.0, 58.0),
    'Excellent': (52.0, 60.0),
    'Very Good': (51.0, 62.0)
}

# Add range categories
df['Depth_Range'] = df['Depth %'].apply(lambda x: get_range_category(x, depth_ranges))
df['Table_Range'] = df['Table %'].apply(lambda x: get_range_category(x, table_ranges))

# Create heatmap of average prices by depth and table ranges
pivot_price = df.pivot_table(
    values='Price',
    index='Depth_Range',
    columns='Table_Range',
    aggfunc='mean'
).round(2)

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_price, annot=True, fmt='.0f', cmap='YlOrRd')
plt.title('Average Price by Depth and Table Ranges')
plt.tight_layout()
plt.show()

# Calculate correlation matrix
correlation_matrix = df[['Price', 'Depth %', 'Table %']].corr()
print("\nCorrelation Matrix:")
print(correlation_matrix.round(3))

# Analyze price premium for optimal proportions
ideal_props = df[
    (df['Depth_Range'] == 'Ideal') & 
    (df['Table_Range'] == 'Ideal')
]

other_props = df[
    (df['Depth_Range'] != 'Ideal') | 
    (df['Table_Range'] != 'Ideal')
]

print("\nPrice Premium Analysis:")
print(f"Average price for ideal proportions: ${ideal_props['Price'].mean():.2f}")
print(f"Average price for other proportions: ${other_props['Price'].mean():.2f}")
premium = ((ideal_props['Price'].mean() / other_props['Price'].mean()) - 1) * 100
print(f"Premium for ideal proportions: {premium:.1f}%")
```

7. Length-to-Width Ratio and Shape Preferences
Analyze how the length-to-width ratio varies across different shapes and whether it correlates with price. Certain shapes (e.g., pear, oval) may have an optimal ratio that maximizes price.

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('diamonds (cleaned).csv')

# Scatter plot of Length/Width Ratio vs Price by Shape
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Length/Width Ratio', y='Price', hue='Shape', alpha=0.6)
plt.title('Length/Width Ratio vs Price by Shape')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Box plot of Length/Width Ratio by Shape
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Shape', y='Length/Width Ratio')
plt.xticks(rotation=45)
plt.title('Length/Width Ratio Distribution by Shape')
plt.tight_layout()
plt.show()

# Calculate optimal ratios (where price is highest) for each shape
optimal_ratios = df.groupby('Shape').apply(lambda x: x.loc[x['Price'].idxmax(), 'Length/Width Ratio']).round(2)
avg_ratios = df.groupby('Shape')['Length/Width Ratio'].mean().round(2)
price_correlation = df.groupby('Shape').apply(lambda x: x['Length/Width Ratio'].corr(x['Price'])).round(3)

results = pd.DataFrame({
    'Optimal Ratio': optimal_ratios,
    'Average Ratio': avg_ratios,
    'Price Correlation': price_correlation
}).sort_values('Price Correlation', ascending=False)

print("\nShape Analysis:")
print(results)
```
